```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
randn = np.random.randn
```

```python
ts = pd.Series(randn(1000), index = pd.date_range('2000/1/1', periods = 1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot(style = '<--')
```

```python
pd.rolling_mean(ts,60).plot(style = '--' ,c = 'r')
```

```python
pd.rolling_mean(ts,180).plot(style = '--' ,c = 'r')
```

```python
plt.show()
```
